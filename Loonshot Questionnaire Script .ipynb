```python
def chunks_generator(my_list, chunk_size):
    """Yield successive n-sized chunks from list my_list."""
    for i in range(0, len(my_list), chunk_size):
        yield my_list[i:i + chunk_size]
```

```python
#Import data in Loonshot_Questionnaire table
def insert_rows_presto(df,cnxn,presto_table):

    df_tmp_control = df.copy()
    for col in df_tmp_control.columns:
        df[col] = df_tmp_control[col]
        
    table = presto_table
    print('Creating BI table...')
    print(' ')
    rows = 0
    chunk_rows = 500
    length_df = len(df)
    for data in chunks_generator([str(tuple(row)) for row in df.values], chunk_rows):
        rows = rows + chunk_rows
        print(f'Inserted {rows} out of {length_df} rows')
        data = map(lambda el: el.replace('"', ''), data)
        query = "INSERT INTO " + table + " VALUES " + ", ".join(data) 
#         print('')
#         print(query)
#         print('')
        pd.read_sql_query(query,cnxn)
    print('')
    print(f'Finished creating {presto_table} table')
    print('')
```

```python
insert_rows_presto(Questionnaire_df,cnxn,'loonshot_questionnaire')
```
